```python
from tensorflow.keras.models import load_model

model = load_model('best_model (5).keras')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Menggunakan ImageDataGenerator untuk data uji
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    '/content/belens-dataset/Minuman Dataset',  # Ganti dengan path data uji Anda
    target_size=(120, 120),  # Ukuran input untuk model
    batch_size=32,
    class_mode='categorical',
    shuffle=False  # Agar urutan data tetap
)

predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)  # Mengambil kelas dengan probabilitas tertinggi

# Mendapatkan label sebenarnya (true labels)
true_classes = test_generator.classes

# Prediksi model
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)

# Mendapatkan label sebenarnya (true labels)
true_classes = test_generator.classes

# Menghitung classification report
report = classification_report(true_classes, predicted_classes, target_names=test_generator.class_indices.keys())

# Menampilkan classification report
print(report)
```
